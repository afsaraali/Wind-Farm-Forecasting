```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train.set_index('date', inplace=True)
```

```python
# plot gaps in the time series data

# resample so missing hours are filled with NaN (including the gap after the series finished)
ix = pd.date_range(start=train.index[0], end=train.index[-1]+pd.Timedelta(hours=48), freq='H')
no_gaps_df = train.copy().reindex(ix)

# create boolean for gap or no gap
gaps = no_gaps_df.isna().astype(int)

# plot gaps
fig, ax = plt.subplots(7, 1, figsize=(12, 10))
for n in range(7):
    column = 'wp'+str(n + 1)
    ax[n].plot(gaps.index, gaps[column], lw = 1)
    ax[n].set_title('Power ouput - wind farm '+str(n + 1))
plt.tight_layout();
```

```python
# programatically verify correct pattern in test data

# the first data gap is on 01/01/2011
# split 36 hours prior to this to include the available power data before the first forecasting period 
test_split_date = datetime(2011,1,1,1) - pd.Timedelta(hours = 36) 
power_test_data = train.loc[test_split_date:,:]

# find number of repeated patterns expected in hold-out set - note we expect two per week
# note that the last 48 hours of missing data is allowed for
num_days = (power_test_data.index[-1]+pd.Timedelta(hours=48)) - (power_test_data.index[0]-pd.Timedelta(hours=1))
num_patterns = int(num_days/pd.Timedelta(days=3, hours=12))

# write expected pattern to list - repeating pattern of 36 hour period of data, then 48 period without data
expected_pattern = ([0]*36+[1]*48)*num_patterns

# test number of unexpected gaps in hold out test set
ext_test_gaps = gaps.loc[test_split_date:,:]
print(ext_test_gaps)
```

```python
# Using 16 weeks of testing data (~20% of full training set)
val_split_date = test_split_date - pd.Timedelta(weeks = 16) 

power_train = train.loc[:val_split_date - pd.Timedelta(hours = 1), :] # using inclusive indexing
power_val = train.loc[val_split_date:test_split_date - pd.Timedelta(hours = 1),:]

# Finding number of repeated patterns in validation set (36 hours of observed data + 48 hours of missing data)
num_patterns = int(power_val.shape[0]/84)
```

```python
# create index list for data and target datasets

data_index = []
target_index = []

for pattern_num in range(num_patterns):
    data_index.extend(list(range(84*pattern_num, 84*pattern_num+36)))
    target_index.extend(list(range(84*pattern_num+36, 84*pattern_num+84)))
```

```python
# create data and target dataframes
power_val_data = power_val.iloc[data_index,:]
power_val_target = power_val.iloc[target_index,:]
```

```python
# create benchmark model based on persistent forecast method
# resample data using forward filling to get persistent forecast
power_val_benchmark = power_val_data.reindex(power_val.index, method='ffill')
# reduce to just forecasted data
power_val_benchmark = power_val_benchmark.iloc[target_index,:]
```

```python
power_train.to_csv('../datasets/power_train.csv')
power_val_data.to_csv('../datasets/power_val_data.csv')
power_val_target.to_csv('../datasets/power_val_target.csv')
power_val_benchmark.to_csv('../datasets/power_val_benchmark.csv')
power_test_data.to_csv('../datasets/power_test_data.csv')
```
